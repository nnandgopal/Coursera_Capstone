# Data gathering and cleaning

```python
import pandas as pd # library for data analsysis
```

```python
toronto = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M") #Scraping the wiki page
df_can = toronto[0]# Getting the table from the wiki page as a dataframe
```

```python
df_can.shape
df_can.head()
```

```python
df_can = df_can[df_can['Borough'] !='Not assigned'].reset_index(drop=True, ) #processing only the cells that have an assigned borough. 
df_can.shape
```

```python
df_can[df_can.duplicated('Postcode')]
df_can[df_can['Postcode']=='M9B']
```

```python
df_can =df_can.groupby(['Postcode','Borough'])
df_can=df_can['Neighbourhood'].apply(', '.join).reset_index() 
df_can.head()
```

```python
for n in range(len(df_can)):
    if df_can['Neighbourhood'].iloc[n] =='Not assigned':
        df_can['Neighbourhood'].iloc[n]= df_can['Borough'].iloc[n]
```

```python
df_can.rename(columns={"Postcode": "PostalCode"}, inplace = True)
df_can
```

```python
df_can.shape
```

```python
df_tor = pd.read_csv("http://cocl.us/Geospatial_data")
df_tor.rename(columns={"Postal Code": "PostalCode"},inplace = True)
df_tor
```

# Merging the tables 

```python
df_tor1 = pd.merge(df_tor, df_can, on='PostalCode', how = 'outer')  #combine dataframes

df_tor1 
```

```python
df_toronto = df_tor1[['PostalCode', 'Borough', 'Neighbourhood', 'Latitude', 'Longitude']]  #rearrange data frame

df_toronto
```

## The next code block is to identify the latitude and longiture of Toronto

```python
address = 'Toronto, Ontario'

geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinates of Toronto are {}, {}.'.format(latitude, longitude))
```

## This will add the co-ordinates of the neighbourhoods to the map

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_toronto['Latitude'], df_toronto['Longitude'], df_toronto['Borough'], df_toronto['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
